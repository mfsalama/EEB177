```python
# read in the file and make dictionaries of DBH at each census

#in_file = open("/home/eeb177-student/Desktop/eeb-177/lab-work/exercise-7/cocoli.txt")
in_file = open("/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-21-Feb/cocoli.txt")
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}
#get rid of header row
in_file.readline()

# work on all subsequent
for line in in_file:
    items = line.split()
    tree_id = items[0]
    #retrieve relevant columns
    s94 = float(items[4])
    s97 = float(items[5])
    s98 = float(items[6])
    
    #populate our dictionaries
    sizes_94[tree_id] = s94
    sizes_97[tree_id] = s97
    sizes_98[tree_id] = s98
    
in_file.close()
```

```python
import matplotlib.pyplot as plt

# lets make a hisitogram of DBHs in 1994
# matplotlib works on lists, so lets make one
sizes_in_1994 = list(sizes_94.values())
```

```python
plt.hist(sizes_in_1994)
plt.show()
```

```python
sizes_in_1994 = list(sizes_94.values())
sizes_in_1998 = list(sizes_98.values())

f, (ax1, ax2) = plt.subplots(1,2,sharex = True, sharey = True)
ax1.hist(sizes_in_1994, log = True)
ax1.set_title('Hist in 1994')
ax2.hist(sizes_in_1998, log = True)
ax2.set_title('Hist in 1998')
plt.show()
```
